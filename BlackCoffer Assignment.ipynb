```python
import openpyxl
import requests
from bs4 import BeautifulSoup
import os

# Load the input Excel file
input_workbook = openpyxl.load_workbook('input.xlsx')
input_worksheet = input_workbook.active

# Create a directory to save the extracted articles
output_dir = 'extracted_articles'
os.makedirs(output_dir, exist_ok=True)



# Iterate through the rows of the input worksheet
for row in input_worksheet.iter_rows(min_row=2, max_row=input_worksheet.max_row, min_col=1, max_col=2):
    article_name = row[0].value
    article_url = row[1].value

    if article_name and article_url:
        # Fetch the content of the URL
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
        response = requests.get(article_url, headers=headers)
        content = response.content

        # Parse the content using BeautifulSoup
        soup = BeautifulSoup(content, 'html.parser')
   
        # Extract the text from the specified elements
        h1_title = ""
        if soup.find('h1', class_='entry-title'):
            h1_title = soup.find('h1', class_='entry-title').get_text()
        elif soup.find('h1', class_='tdb-title-text'):
            h1_title = soup.find('h1', class_='tdb-title-text').get_text()
            
        div_content = ""

        if soup.find('div', class_='td-post-content tagdiv-type'):
            div_content = soup.find('div', class_='td-post-content tagdiv-type').get_text()
        elif soup.find('div', class_='tdb-block-inner td-fix-index'):
            target_divs = soup.find_all('div', class_='tdb-block-inner td-fix-index')
            largest_div = None
            largest_length = 0

            # Iterate through the target <div> elements
            for div in target_divs:
                div_text = div.get_text()
                div_length = len(div_text)

                # Compare the lengths to find the largest div
                if div_length > largest_length:
                    largest_div = div
                    largest_length = div_length

            # Print the content of the largest div
            if largest_div:
                div_content=largest_div.get_text()
       
        
        # Combine the extracted texts
        article_text = h1_title + '\n\n' + div_content

        # Create a text file and save the article text
        file_path = os.path.join(output_dir, f'{article_name}.txt')
        with open(file_path, 'w', encoding='utf-8') as file:
            file.write(article_text)

# Close the input workbook
input_workbook.close()
```

```python
import re
import statistics
from collections import Counter
extracted_articles_folder = "extracted_articles"
output_folder = "processed_articles"

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

dic = pyphen.Pyphen(lang='en')

personal_pronouns = ["i", "we", "my", "ours", "us"]

for filename in os.listdir(extracted_articles_folder):
    #print(filename)
    if filename.endswith(".txt"):
        with open(os.path.join(extracted_articles_folder, filename), "r", encoding="utf-8") as article_file:
            content = article_file.read().replace("\n", " ")
            
            sentences = sent_tokenize(content)

            # Calculate average sentence length
            total_sentence_length = sum(len(re.findall(r'\b\w+\b', sentence)) for sentence in sentences)
            average_sentence_length = total_sentence_length / len(sentences) if sentences else 0
            
            words = re.findall(r'\b\w+\b', content.lower())  # Tokenize words

            cleaned_words = [word for word in words if word not in stop_words]

            positive_word_count = sum(1 for word in cleaned_words if word in positive_words)
            negative_word_count = sum(1 for word in cleaned_words if word in negative_words)


            polarity_score = (positive_word_count - negative_word_count) / (positive_word_count + negative_word_count + 1e-6)
            subjectivity_score = (positive_word_count + negative_word_count) / (len(cleaned_words) + 1e-6)

            sentiment = "Positive" if polarity_score > 0 else "Negative" if polarity_score < 0 else "Neutral"
            
            # Calculate the percentage of complex words (words with more than two syllables)
            complex_word_count = sum(1 for word in cleaned_words if len(dic.inserted(word).split("-")) > 2)
            complex_word_percentage = (complex_word_count / len(cleaned_words)) * 100 if len(cleaned_words) > 0 else 0

            # Calculate the Gunning Fog Index
            fog_index = 0.4 * (average_sentence_length + complex_word_percentage)
            
             # Calculate the average number of words per sentence
            total_words = len(words)
            total_sentences = len(sentences)
            average_words_per_sentence = total_words / total_sentences if total_sentences > 0 else 0
            
            # Calculate the syllable count per word
            syllable_counts = []
            for word in cleaned_words:
                if not word.endswith('es') and not word.endswith('ed'):
                    syllable_count = len(dic.inserted(word).split("-"))
                    syllable_counts.append(syllable_count)
            
             # Count personal pronouns
            personal_pronoun_count = [word for word in words if word in personal_pronouns]
            
            # Calculate the average word length
            total_characters = sum(len(word) for word in cleaned_words)
            average_word_length = total_characters / len(cleaned_words) if len(cleaned_words) > 0 else 0

        
            
            with open(os.path.join(output_folder, filename), "w", encoding="utf-8") as output_file:
                output_file.write(f"POSITIVE SCORE: {positive_word_count:}\n")
                output_file.write(f"NEGATIVE SCORE: {negative_word_count:}\n")
                output_file.write(f"POLARITY SCORE: {polarity_score:.4f}\n")
                output_file.write(f"SUBJECTIVITY SCORE: {subjectivity_score:.4f}\n")
                output_file.write(f"AVG SENTENCE LENGTH: {average_sentence_length:.0f}\n")
                output_file.write(f"PERCENTAGE OF COMPLEX WORDS: {complex_word_percentage:.2f}%\n")
                output_file.write(f"FOG INDEX: {fog_index:.2f}\n")
                output_file.write(f"AVG NUMBER OF WORDS PER SENTENCE: {average_words_per_sentence:.0f}\n")
                output_file.write(f"COMPLEX WORD COUNT: {complex_word_count:}\n")
                output_file.write(f"WORD COUNT: {len(cleaned_words):}\n")
                output_file.write(f"SYLLABLE PER WORD: {statistics.mean(syllable_counts):.0f}\n")
                output_file.write(f"PERSONAL PRONOUNS: {len(personal_pronoun_count)}\n")
                output_file.write(f"AVG WORD LENGTH: {average_word_length:.0f}\n")
               
```
